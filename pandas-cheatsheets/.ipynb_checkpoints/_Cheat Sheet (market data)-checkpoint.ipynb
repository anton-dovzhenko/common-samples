```python
from datetime import datetime
from itertools import chain, cycle
import numpy as np
import pandas as pd
from typing import List

from IPython.display import display, HTML, display_html
display(HTML("<style>.container { width:1300px !important; }</style>"))


def display_side_by_side(*args,titles=cycle([''])):
    html = ''
    for df, title in zip(args, chain(titles, cycle(['</br>'])) ):
        html += '<th style="text-align:center"><td style="vertical-align:top">'
        html += f'<h4>{title}</h4>'
        html += df.to_html().replace('table','table style="display:inline"')
        html += '</td></th>'
    display_html(html, raw=True)
```

### Market data generation    
- negative level stands for bid, positive levels - for ask

```python
def get_market_mids(start_rate, start_time, n, std, period):
    rate = start_rate * np.cumprod(np.random.normal(1, std, n))
    index = start_time + period[0] * pd.to_timedelta(np.arange(n), period[1])
    return pd.Series(rate, index=index, name='mid')


def get_market_data(market, mids, period, rungs, precision=5):
    index = pd.date_range(start=mids.index[0], end=mids.index[-1], freq=period)
    N = len(index)
    R = len(rungs)
    data = pd.Series(np.arange(N), index=index, name='mock')
    data = pd.merge_asof(data, mids, direction='backward', left_index = True, right_index = True)
    data['size'] = [rungs] * N
    
    def get_side(side_data, levels, price_formula):
        side_data['level'] = [levels] * N
        side_data['spread'] = np.sort(1e-4 * np.random.random((N, R))).tolist()
        side_data = side_data.explode(['spread', 'size', 'level'])
        side_data[['spread', 'size', 'level']] = side_data[['spread', 'size', 'level']].apply(pd.to_numeric)
        side_data['price'] = np.around(side_data.eval(price_formula), decimals=precision)
        return side_data
    
    bid = get_side(data.copy(), list(range(-1, -(R + 1), -1)), 'mid * (1 - spread)')
    ask = get_side(data.copy(), list(range(1, R + 1)), 'mid * (1 + spread)')
    data = pd.concat([bid, ask])
    data.drop(['mock', 'mid', 'spread'], axis=1, inplace=True)
    data['market'] = market
    data.sort_index(inplace=True)
    return data
```

```python
static_mids = get_market_mids(1.05, datetime(2022, 1, 2), 1000000, 0.00005, (5, 'ms'))
static_mids.plot()
static_mids.head()
```

```python
md_a = get_market_data('Aa', static_mids, '101ms', [1e6, 2e6, 5e6])
md_b = get_market_data('Bb', static_mids, '53ms', [1e6, 2e6])
md_c = get_market_data('Cc', static_mids, '184ms', [1e6, 5e6])
```

```python
display_side_by_side(md_a.head(12), md_b.head(12), titles=['Market Aa', 'Market Bb'])
```

### Market data functions
- [+] build order book
- [+] get TOB (single market)
- [+] get TOB (order book)
- [+] get VWAP (order book)
- [+] get time-weighted avg spread

#### Build order book

```python
def build_order_book(market_stacks : List[pd.DataFrame]):
    # get update times from all markets in idx
    idx = np.concatenate(list(x.index.values for x in market_stacks))
    idx = np.sort(np.unique(idx))
    idx = pd.DataFrame({'time': idx, 'mock': range(len(idx))})
    
    # capture last tick from every market for any market update
    ob = []
    for ms in market_stacks:
        ms = ms.reset_index()
        ms = ms.rename(columns={'index': 'tick_time'})
        time = pd.merge_asof(idx, ms[['tick_time']], direction='backward',  left_on='time', right_on='tick_time')
        time.drop('mock', axis=1, inplace=True)
        ms = pd.merge(ms, time, on='tick_time')
        ob.append(ms)
    ob = pd.concat(ob).sort_values(by=['time', 'market', 'level'])
    
    # group bids and asks separately
    ob_grouped_bid = ob[ob['level'] < 0].groupby('time')
    ob_grouped_ask = ob[ob['level'] > 0].groupby('time')
    ob = ob_grouped_bid['price'].apply(np.array).to_frame(name='bid')
    ob['bidSize'] = ob_grouped_bid['size'].apply(np.array).values
    ob['bidMarket'] = ob_grouped_bid['market'].apply(np.array).values
    ob['ask'] = ob_grouped_ask['price'].apply(np.array).values
    ob['askSize'] = ob_grouped_ask['size'].apply(np.array).values
    ob['askMarket'] = ob_grouped_ask['market'].apply(np.array).values
    
    # order bids (desc) and asks (asc)
    ob['bidOrder'] = ob['bid'].apply(lambda x: np.argsort(-x))
    ob['askOrder'] = ob['ask'].apply(lambda x: np.argsort(x))
    order = lambda x: x[0][x[1]]
    ob['bid'] = ob[['bid', 'bidOrder']].apply(order, axis=1)
    ob['bidSize'] = ob[['bidSize', 'bidOrder']].apply(order, axis=1)
    ob['bidMarket'] = ob[['bidMarket', 'bidOrder']].apply(order, axis=1)
    ob['ask'] = ob[['ask', 'askOrder']].apply(order, axis=1)
    ob['askSize'] = ob[['askSize', 'askOrder']].apply(order, axis=1)
    ob['askMarket'] = ob[['askMarket', 'askOrder']].apply(order, axis=1)
    
    ob.drop(['bidOrder', 'askOrder'], axis=1, inplace=True)
    
    return ob
```

```python
build_order_book([md_a, md_b, md_c])[:5]
```

#### Get TOB (from single market)

```python
def get_tob_from_single_market(md):
    bid = md[md['level'] == -1][['price', 'size']].rename(columns={'price': 'bid', 'size': 'bidSize'})
    ask = md[md['level'] == 1][['price', 'size']].rename(columns={'price': 'ask', 'size': 'askSize'})
    bid['bidTime'] = bid.index.values
    ask['askTime'] = ask.index.values
    result = pd.DataFrame({'time': np.unique(np.sort(md.index.values))})
    result = pd.merge_asof(result, bid, direction='backward', left_on='time', right_index=True)
    result = pd.merge_asof(result, ask, direction='backward', left_on='time', right_index=True)
    result.set_index('time', inplace=True)
    return result
```

```python
get_tob_from_single_market(md_a)[:5]
```

#### Get TOB (from order book)

```python
def get_tob_from_order_book(ob):
    cols = ['bid', 'bidSize', 'bidMarket', 'ask', 'askSize', 'askMarket']
    cols = list(set(cols) & set(ob.columns))
    for c in cols:
        ob[c] = ob[c].apply(lambda x: None if 0 == len(x) else x[0])
    return ob
```

```python
get_tob_from_order_book(build_order_book([md_a[:100], md_b[:100], md_c[:100]]))
```

#### Get VWAP (from order book)

```python
def get_vwap_from_order_book(ob, amount):
    get_weights = lambda amount, size: np.diff(size.cumsum().clip(0, amount), prepend=0.0)
    wavg = lambda x: np.average(x[0], weights=get_weights(amount, x[1]))
    ob['bidVwap'] = ob[['bid', 'bidSize']].apply(wavg, axis=1)
    ob['askVwap'] = ob[['ask', 'askSize']].apply(wavg, axis=1)
    return ob[['bidVwap', 'askVwap']].rename(columns={'bidVwap': 'bid', 'askVwap': 'ask'})
```

```python
vwap_2_5 = get_vwap_from_order_book(build_order_book([md_a[:100], md_b[:100], md_c[:100]]), 2.5e6)
vwap_5 = get_vwap_from_order_book(build_order_book([md_a[:100], md_b[:100], md_c[:100]]), 5e6)
display_side_by_side(vwap_2_5.head(), vwap_5.head(), titles=['2.5mil', '5mil'])
```

#### Get time-weighted average spread

```python
def get_time_weighted_spread(data, rule):
    # FIXME: ignore NaN times and spreads for first and last entries
    data['spread'] = data.eval('ask - bid')
    data['tick_time'] = data.index
    data = data[['tick_time', 'spread']].resample(rule).apply(list)
    
    data['open_spread'] = data['spread'].apply(lambda x: None if 0 == len(x) else x[-1]).shift(periods=1)
    data['spread'] = data[['open_spread', 'spread']].apply(lambda x: np.array([x[0]] + x[1]), axis=1)
    
    data['open'] = data.index.values
    data['close'] = data['open'].shift(periods=-1)
    data['tick_time'] = data[['tick_time', 'open', 'close']].apply(lambda x: np.diff(np.array([x[1]] + x[0] + [x[2]])), axis=1)
    data['spread'] = data[['spread', 'tick_time']].apply(lambda x: np.average(x[0], weights=x[1]), axis=1)
    return data[['spread']]
```

```python
get_time_weighted_spread(get_tob_from_single_market(md_a), '60s').plot()
```
